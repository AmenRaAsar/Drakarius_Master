```python
# ===============================================
# 🚀 DRAKARIUS MASTER SIMULATION — COLAB EDITION
# ===============================================

# === 1️⃣ Mount Google Drive (for persistence)
from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive
!mkdir -p Drakarius_Simulation && cd Drakarius_Simulation
%cd Drakarius_Simulation

# === 2️⃣ System Info
!nvidia-smi || echo "⚠️ No GPU found - running CPU mode"
!lscpu | grep "Model name"

# === 3️⃣ Install Dependencies
!apt update -y
!apt install -y build-essential cmake gfortran libopenmpi-dev python3-dev
!pip install numpy scipy matplotlib sympy torch fenics h5py tqdm

# === 4️⃣ Import Core Libraries
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from scipy.constants import c, epsilon_0, mu_0
import torch

# === 5️⃣ Define Simulation Parameters
grid_size = 256       # spatial resolution
dx = 1e-6             # spatial step (m)
dt = dx / (2*c)       # time step
timesteps = 500       # number of steps

# Beetle-inspired lattice refractive index modulation
def beetle_lattice(x, y):
    return 1.5 + 0.3*np.sin(8*np.pi*x)*np.cos(8*np.pi*y)

# === 6️⃣ Initialize Fields
E = np.zeros((grid_size, grid_size))
H = np.zeros_like(E)
S = np.zeros_like(E)

# === 7️⃣ Source parameters (Drakarius resonator)
freq = 5e14  # laser frequency (Hz)
omega = 2*np.pi*freq
amplitude = 1.0

# === 8️⃣ Run Field Propagation Loop
for t in tqdm(range(timesteps)):
    E = amplitude * np.sin(omega*t*dt) * beetle_lattice(
        *np.meshgrid(np.linspace(0,1,grid_size), np.linspace(0,1,grid_size))
    )
    H = E / 377.0
    S = (E**2) / 377.0  # Poynting vector magnitude approximation

# === 9️⃣ Compute Physics Metrics
U = 0.5 * epsilon_0 * E**2  # energy density
mean_S = np.mean(S)
max_S = np.max(S)
mean_U = np.mean(U)
max_U = np.max(U)

print(f"Poynting Vector Mean: {mean_S:.3e} W/m²")
print(f"Poynting Vector Max: {max_S:.3e} W/m²")
print(f"Energy Density Mean: {mean_U:.3e} J/m³")
print(f"Energy Density Max: {max_U:.3e} J/m³")

# === 🔢 Derived Metrics
pressure = mean_S / c
momentum_flux = pressure * 4 * np.pi * (1e-6)**2

print(f"Field Pressure: {pressure:.3e} N/m²")
print(f"Momentum Flux (approx recoil): {momentum_flux:.3e} N")

# === 🔬 Visualization
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(S, cmap='inferno')
plt.title("Poynting Vector Magnitude (W/m²)")
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(U, cmap='plasma')
plt.title("Energy Density (J/m³)")
plt.colorbar()
plt.suptitle("Drakarius Beetle-Lattice Resonator Field Distribution")
plt.tight_layout()
plt.savefig("Drakarius_FieldMap.png", dpi=200)
plt.show()

# === 💾 Save Results
np.savetxt("Drakarius_Poynting.csv", S, delimiter=",")
np.savetxt("Drakarius_Energy.csv", U, delimiter=",")
with open("Drakarius_Results.txt", "w") as f:
    f.write(f"P_mean={mean_S:.3e}, P_max={max_S:.3e}, U_mean={mean_U:.3e}, U_max={max_U:.3e}\n")

print("✅ Simulation complete — results saved to Google Drive /Drakarius_Simulation/")
```
